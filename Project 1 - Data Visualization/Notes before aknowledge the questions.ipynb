```python
Movies = Movies.drop(Movies.loc[Movies['unknown'] == 1].index)
```

<font size=6>Number of Movies per Genre</font>

```python
MoviesGenre = Movies.drop(['movie id', 'movie title', 'release date', 'unknown'], axis = 1).sum().sort_values()
```

```python
mx = MoviesGenre.index
my = MoviesGenre.values

plt.rcParams["figure.figsize"] = (8,6)
plt.barh(mx, my)

for index, value in enumerate(my):
    plt.text(value, index, str(value))
```

<font size=6>Number Of Movies having more than 1 Genre</size>

```python
Movies["type count"] = Movies.loc[:,'unknown':'Western'].sum(axis=1)
print("Movies with more than 1 genre: " + str(len(Movies.loc[Movies['type count'] > 1])))
```

<font size=6>Genre Populatiry</font>

```python
Movies['release date'] = pd.to_datetime(Movies['release date'])
Movies['release year'] = Movies['release date'].dt.year
```

```python
OneForAllMR = pd.merge(left=Movies, right=UsersRating , on="movie id",  how='left')
OneForAllUS = pd.merge(left=OneForAllMR, right=Users , on="user id",  how='outer')
```

```python
listPopularity = []
for genre in MoviesGenre.index:
    PopularityHeat = pd.DataFrame(OneForAllMR, columns=['release year', 'rating']).loc[OneForAllMR[genre] != 0].sort_values('release year')
    PopularityHeat = PopularityHeat.groupby('release year').mean()
    PopularityHeat = PopularityHeat.rename(columns={"rating": genre }).agg(list).T
    
    listPopularity.append(PopularityHeat)
    
PopularityHeatJ = pd.concat(listPopularity)
PopularityHeatJ = PopularityHeatJ.fillna(0)
fig,ax = plt.subplots(figsize=(30,10))
sns.heatmap(PopularityHeatJ, ax=ax, annot=True, linewidths=0.05,cmap='magma')
plt.show()
```

```python
for genre in MoviesGenre.index:
    PopularityData = pd.DataFrame(OneForAllMR, columns=['release year', 'rating']).loc[OneForAllMR[genre] != 0].sort_values('release year')
    PopularityG = PopularityData.groupby('release year', as_index=False).mean().agg(list)
    
    plt.rcParams["figure.figsize"] = (5,2)
    plt.plot(PopularityG["release year"], PopularityG["rating"])        
    plt.ylim(top=5)
    plt.ylim(bottom=0)
    plt.xlabel('Year')
    plt.ylabel('average rating')
    plt.title(genre)
    print(plt.show())
    
    plt.rcdefaults()
```

<font size=6>Top 25 Movies</font>

```python
MoviesRating = pd.DataFrame(OneForAllMR, columns=['movie title', 'rating'])
MoviesRatingG = MoviesRating.groupby('movie title', as_index=False).mean().agg(list).sort_values(by='rating', ascending = False)
MoviesRatingG = MoviesRatingG.head(25).sort_values(by='rating')
MoviesRatingG


mox = MoviesRatingG['movie title']
moy = MoviesRatingG['rating']

plt.rcParams["figure.figsize"] = (10,10)

fig, ax = plt.subplots()

ax.barh(mox, moy)
ax.set_xlabel('Mean Rating')

plt.show()
```

<font size=6>Gender Distribution</font>

```python
GeneralGender = []
for genre in MoviesGenre.index:
    GenderDistribution = pd.DataFrame(OneForAllUS, columns=['gender']).loc[OneForAllUS[genre] != 0]
    GenderDistribution[genre] = np.zeros(len(GenderDistribution))
    GenderDistribution = GenderDistribution.groupby('gender').count().T
    
    GeneralGender.append(GenderDistribution)
    
GeneralGenderJ = pd.concat(GeneralGender)
plt.rcParams["figure.figsize"] = (8,6)
grplt = GeneralGenderJ.plot(kind='barh')
grplt.set_xlabel("Reviews count")
grplt.set_ylabel("Genre")

#This is a general contabilization of the reviews by gender
```
